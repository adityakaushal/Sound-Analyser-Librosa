```python
import librosa as lb
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = "/Users/adityakaushal/Desktop/Beautiful Day.mp3"
```

```python
time_series, sample_rate = lb.load(filename)
```

```python
stft = np.abs(lb.stft(time_series, hop_length = 512, n_fft = 2048*4))
```

```python
stft
```

```python
# converting the matrix to decibel matrix
spectrogram = lb.amplitude_to_db(stft, ref = np.max)
```

```python
librosa.display.specshow(spectrogram,y_axis='log', x_axis='time')
plt.title('U2 Beautiful Day')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.show()
```
